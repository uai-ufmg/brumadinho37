```python
import pandas as pd
import numpy as np

df = pd.read_csv("pre_taxa_sih_cap.csv")
df
```

```python
df['id_municipio'] = df['id_municipio'].astype('uint32')
df['ano'] = df['ano'].astype('uint16')
df['pop_total'] = df['pop_total'].astype('uint64')
df['pop_pad_total'] = df['pop_pad_total'].astype('uint64')
df['pop_homem'] = df['pop_homem'].astype('uint64')
df['pop_pad_homem'] = df['pop_pad_homem'].astype('uint64')
df['pop_mulher'] = df['pop_mulher'].astype('uint64')
df['pop_pad_mulher'] = df['pop_pad_mulher'].astype('uint64')
df['mes'] = df['mes'].astype('uint8')
df['inter_total'] = df['inter_total'].astype('uint32')
df['inter_homem'] = df['inter_homem'].astype('uint32')
df['inter_mulher'] = df['inter_mulher'].astype('uint32')
df['taxa_inter_total'] = df['taxa_inter_total'].astype('float32')
df['taxa_inter_homem'] = df['taxa_inter_homem'].astype('float32')
df['taxa_inter_mulher'] = df['taxa_inter_mulher'].astype('float32')
df['inter_esp_total'] = df['inter_esp_total'].astype('float32')
df['inter_esp_homem'] = df['inter_esp_homem'].astype('float32')
df['inter_esp_mulher'] = df['inter_esp_mulher'].astype('float32')
df['grupo'] = df['grupo'].astype('uint8')
df['taxa_inter_esp_total'] = df['taxa_inter_esp_total'].astype('float32')
df['taxa_inter_esp_homem'] = df['taxa_inter_esp_homem'].astype('float32')
df['taxa_inter_esp_mulher'] = df['taxa_inter_esp_mulher'].astype('float32')

#df.to_hdf("pre_taxa_sih.h5", "df")
df.info()
```

```python
df_sf = df.groupby(['municipio', 'id_municipio', 'ano', 'mes', 'capitulo'], as_index=False).sum()
```

```python
#df_total_agravo = df.groupby(['id_municipio','agravo', 'ano', 'mes'], as_index=False)['inter_total'].sum()
df_total = df.groupby(['id_municipio', 'ano', 'mes'], as_index=False)['inter_total'].sum()
```

```python
df_total.columns = ['id_municipio', 'ano', 'mes', 'total_inter_loc']
df_total
```

```python
df_sf = df_sf.merge(df_total, on=['id_municipio', 'ano', 'mes'])
df_sf
```

```python
df_sf['taxa_nova'] = np.nan
df_sf['taxa_nova'] = df_sf['inter_total']/df_sf['pop_total']
df_sf[df_sf['taxa_nova'] > 0]
```
