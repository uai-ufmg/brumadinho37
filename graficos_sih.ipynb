```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt

df = pd.read_csv("pre_taxa_sih.csv")
df
```

```python
pop_pad_t = df['pop_pad_total'].unique().sum()
pop_pad_h = df['pop_pad_homem'].unique().sum()
pop_pad_m = df['pop_pad_mulher'].unique().sum()
```

```python
df_19 = df[df['ano'] == 2019]
pop_19 = df_19.groupby(['agravo', 'grupo'], as_index=False).sum()

pop_19['pop_total'] = pop_19['pop_total'] / 12
pop_19['pop_homem'] = pop_19['pop_homem'] / 12
pop_19['pop_mulher'] = pop_19['pop_mulher'] / 12


pop_19['taxa_inter_total'] = (pop_19['inter_total'] / pop_19['pop_total']) * 100000
pop_19['taxa_inter_homem'] = (pop_19['inter_homem'] / pop_19['pop_homem']) * 100000
pop_19['taxa_inter_mulher'] = (pop_19['inter_mulher'] / pop_19['pop_mulher']) * 100000

pop_19['taxa_inter_esp_total'] = (pop_19['inter_esp_total'] / pop_pad_t) * 100000
pop_19['taxa_inter_esp_homem'] = (pop_19['inter_esp_homem'] / pop_pad_h) * 100000
pop_19['taxa_inter_esp_mulher'] = (pop_19['inter_esp_mulher'] / pop_pad_m) * 100000

pop_19
```

```python
df_r = df[df['ano'] != 2019]
pop_r = df_r.groupby(['agravo', 'grupo'], as_index=False).sum()

pop_r['pop_total'] = pop_r['pop_total'] / 12
pop_r['pop_homem'] = pop_r['pop_homem'] / 12
pop_r['pop_mulher'] = pop_r['pop_mulher'] / 12

pop_r['taxa_inter_total'] = (pop_r['inter_total'] / pop_r['pop_total']) * 100000
pop_r['taxa_inter_homem'] = (pop_r['inter_homem'] / pop_r['pop_homem']) * 100000
pop_r['taxa_inter_mulher'] = (pop_r['inter_mulher'] / pop_r['pop_mulher']) * 100000

pop_r['taxa_inter_esp_total'] = (pop_r['inter_esp_total'] / (pop_pad_t * 9)) * 100000
pop_r['taxa_inter_esp_homem'] = (pop_r['inter_esp_homem'] / (pop_pad_h * 9)) * 100000
pop_r['taxa_inter_esp_mulher'] = (pop_r['inter_esp_mulher'] / (pop_pad_m * 9)) * 100000

pop_r
```

```python
def plot_grafico_dif_old(df1, df2, taxa, titulo, xlabel):
    aux = (df1.groupby(['grupo', 'agravo'])[taxa].
           sum().
           unstack(0))
    aux2 = (df2.groupby(['grupo', 'agravo'])[taxa].
            sum().
            unstack(0))

    fig, ax = plt.subplots(1, 1, figsize=(22, 18))

    ax.set_title(f"{titulo}", fontsize=18)
    ax.set_xlabel(xlabel, fontsize=18)
    ax.set_ylabel("agravo", fontsize=18)

    aux3 = aux - aux2
    
    (aux3.sort_values(ascending=False, by=1.0).
     head(15)[::-1].
     plot.barh(ax=ax, edgecolor="k", width=0.7))

    plt.tick_params(axis='x', which='major', labelsize=18)
    plt.tick_params(axis='y', which='major', labelsize=18)
    plt.axvline()
    plt.legend(fontsize=18, loc='lower right')
    
    return ax, fig
```

```python
def plot_grafico_dif(df1, df2, taxa, titulo, xlabel):
    aux = (df1.groupby(['grupo', 'agravo'])[taxa].
           sum().
           unstack(0))
    aux2 = (df2.groupby(['grupo', 'agravo'])[taxa].
            sum().
            unstack(0))
    
    aux3 = aux - aux2
    
    df =(aux3.sort_values(ascending=False, by=1.0).
         head(15))
    df = df.stack().reset_index().rename(columns={ 0: 'taxa' })
    df['grupo'] = df['grupo'].astype(str)

    c = alt.Chart(df, title=titulo).mark_bar().encode(
        alt.Color('grupo:N', title='Grupo'),
        alt.Y('agravo:N', title='Agravo', sort=None),
        alt.X('taxa:Q', title=xlabel),
        alt.Order('grupo:N', sort='ascending'),
        alt.Tooltip('taxa:Q')
    ).properties(
        width=640,
    ).configure_range(
        category={'scheme': 'set1'}
    )
    
    return c
```

```python
chart_t = plot_grafico_dif(pop_19, pop_r,
                         'taxa_inter_esp_total', 
                         "agravos com maior diferença da taxa bruta de internações para 2019 - Total",
                        "diferença da taxa padronizada total")

chart_h = plot_grafico_dif(pop_19, pop_r,
                         'taxa_inter_esp_homem', 
                         "agravos com maior diferença da taxa bruta de internações para 2019 - Total",
                        "diferença da taxa padronizada masculina")

chart_m = plot_grafico_dif(pop_19, pop_r,
                         'taxa_inter_esp_mulher', 
                         "agravos com maior diferença da taxa bruta de internações para 2019 - Total",
                        "diferença da taxa padronizada feminina")

(chart_t | chart_h | chart_m)#.save('tx_pad.png')
```
