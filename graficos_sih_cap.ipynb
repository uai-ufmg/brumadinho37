```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt

df = pd.read_csv("pre_taxa_sih_cap.csv")
df
```

```python
df.columns
```

```python
df_filtro = df.groupby(['municipio', 'id_municipio', 'grupo', 'ano', 'mes', 'capitulo'], as_index=False).sum()

df_total = df.groupby(['id_municipio', 'grupo', 'ano', 'mes'], as_index=False)[['inter_total', 'inter_homem', 'inter_mulher']].sum()
df_total.columns = ['id_municipio', 'grupo', 'ano', 'mes', 'total_inter_loc', 'homem_inter_loc', 'mulher_inter_loc']

df_total = df_filtro.merge(df_total, on=['id_municipio', 'grupo', 'ano', 'mes'])
df_total['taxa_nova_total'] = np.nan
df_total['taxa_nova_homem'] = np.nan
df_total['taxa_nova_mulher'] = np.nan

df_total['taxa_nova_total'] = (df_total['inter_total']/df_total['pop_total'])*100000
df_total['taxa_nova_homem'] = (df_total['inter_homem']/df_total['pop_homem'])*100000
df_total['taxa_nova_mulher'] = (df_total['inter_mulher']/df_total['pop_mulher'])*100000

#df_total[df_total['taxa_nova'] > 0]
```

```python
df = df_total
```

```python
df_19 = df[df['ano'] == 2019]
df_19 = df_19.groupby(['capitulo', 'grupo'], as_index=False).sum()
```

```python
df_r = df[df['ano'] != 2019]
df_r = df_r.groupby(['ano', 'capitulo', 'grupo'], as_index=False).sum()
df_r = df_r.groupby(['capitulo', 'grupo'], as_index=False).mean()
df_r
```

```python
def plot_grafico_dif(df1, df2, taxa, titulo, xlabel):
    aux = (df1.groupby(['grupo', 'capitulo'])[taxa].
           sum().
           unstack(0))
    aux2 = (df2.groupby(['grupo', 'capitulo'])[taxa].
            sum().
            unstack(0))
    
    aux3 = aux - aux2
    
    df = (aux3.sort_values(ascending=False, by=1.0).
         head(15))
    df = df.stack().reset_index().rename(columns={ 0: 'taxa' })
    df['grupo'] = df['grupo'].astype(str)

    c = alt.Chart(df, title=titulo).mark_bar().encode(
        alt.Color('grupo:N', title='Grupo'),
        alt.Y('capitulo:N', title='Capítulo', sort=None),
        alt.X('taxa:Q', title=xlabel),
        alt.Order('grupo:N', sort='ascending'),
        alt.Tooltip('taxa:Q')
    ).properties(
        width=640,
    ).configure_range(
        category={'scheme': 'set1'}
    )
    
    return c
```

```python
chart = plot_grafico_dif(df_19, df_r,
                         'inter_total', 
                         "Capítulos com maior diferença absoluta de internações para 2019 - Total",
                        "Diferença absoluta de internações")

#chart.save('tx_bruta_t.svg')
chart
```

```python
chart = plot_grafico_dif(df_19, df_r,
                         'inter_homem', 
                         "Capítulos com maior diferença absoluta de internações para 2019 - Homens",
                        "Diferença absoluta de internações masculina")

#chart.save('tx_bruta_h.svg')
chart
```
