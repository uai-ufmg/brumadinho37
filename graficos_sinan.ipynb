```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt

df = pd.read_csv("pre_taxa_sinan.csv")

df = df[~df['agravo'].isin(['intoxicacao', 'sindrome', 'chagas', 'colera', 'febreamarela'])]
         
df.loc[df['agravo'] == 'homicidio', 'agravo'] = 'violencia'

df
```

```python
df['agravo'].unique()
```

```python
df_19 = df[df['ano'] == 2019]
df_19 = df_19.groupby(['agravo', 'grupo'], as_index=False).sum()

df_19['pop_total'] = df_19['pop_total'] / 12
df_19['pop_homem'] = df_19['pop_homem'] / 12
df_19['pop_mulher'] = df_19['pop_mulher'] / 12

df_19['taxa_nova_total'] = np.nan
df_19['taxa_nova_homem'] = np.nan
df_19['taxa_nova_mulher'] = np.nan

df_19['taxa_nova_total'] = (df_19['notificacoes_total']/df_19['pop_total'])*100000
df_19['taxa_nova_homem'] = (df_19['notificacoes_homem']/df_19['pop_homem'])*100000
df_19['taxa_nova_mulher'] = (df_19['notificacoes_mulher']/df_19['pop_mulher'])*100000

df_19
```

```python
df_r = df[df['ano'] != 2019]
df_r = df_r.groupby(['agravo', 'grupo'], as_index=False).sum()

df_r['pop_total'] = df_r['pop_total'] / 12
df_r['pop_homem'] = df_r['pop_homem'] / 12
df_r['pop_mulher'] = df_r['pop_mulher'] / 12

df_r['taxa_nova_total'] = np.nan
df_r['taxa_nova_homem'] = np.nan
df_r['taxa_nova_mulher'] = np.nan

df_r['taxa_nova_total'] = (df_r['notificacoes_total']/df_r['pop_total'])*100000
df_r['taxa_nova_homem'] = (df_r['notificacoes_homem']/df_r['pop_homem'])*100000
df_r['taxa_nova_mulher'] = (df_r['notificacoes_mulher']/df_r['pop_mulher'])*100000

df_r
```

```python
aux = (df_19.groupby(['grupo', 'agravo'])['taxa_nova_total'].
        sum().
        unstack(0))

aux2 = (df_r.groupby(['grupo', 'agravo'])['taxa_nova_total'].
        sum().
        unstack(0))
    
#(aux - aux2).sort_values(ascending=False, by=1.0)
df_19.sort_values(ascending=False, by='notificacoes_total')
```

```python
def plot_grafico_dif(df1, df2, taxa, titulo, xlabel):
    aux = (df1.groupby(['grupo', 'agravo'])[taxa].
           sum().
           unstack(0))
    aux2 = (df2.groupby(['grupo', 'agravo'])[taxa].
            sum().
            unstack(0))
    
    aux3 = aux - aux2
    
    df = (aux3.sort_values(ascending=False, by=1.0).
         head(15))
    df = df.stack().reset_index().rename(columns={ 0: 'taxa' })
    df['grupo'] = df['grupo'].astype(str)

    c = alt.Chart(df, title=titulo).mark_bar().encode(
        alt.Color('grupo:N', title='Grupo'),
        alt.Y('agravo:N', title='Agravo', sort=None),
        alt.X('taxa:Q', title=xlabel, scale=alt.Scale(domain=[-160, 10000])),
        alt.Order('grupo:N', sort='ascending'),
        alt.Tooltip('taxa:Q')
    ).properties(
        width=640,
    ).configure_range(
        category={'scheme': 'set1'}
    )
    
    return c
```

```python
chart = plot_grafico_dif(df_19, df_r,
                         'taxa_nova_total', 
                         "Agravos com maior direfença da razão internações/população entre 2019 e o resto do período - Total",
                        "Diferença da Razão internações/população (a cada 100.000 habitantes)")

#chart.save('tx_bruta_t.svg')
chart
```
