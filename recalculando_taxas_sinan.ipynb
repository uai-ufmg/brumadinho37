```python
import pandas as pd
import numpy as np

df = pd.read_csv("pre_taxa_sinan.csv")
df
```

```python
df['id_municipio'] = df['id_municipio'].astype('uint32')
df['ano'] = df['ano'].astype('uint16')
df['pop_total'] = df['pop_total'].astype('uint64')
df['pop_pad_total'] = df['pop_pad_total'].astype('uint64')
df['pop_homem'] = df['pop_homem'].astype('uint64')
df['pop_pad_homem'] = df['pop_pad_homem'].astype('uint64')
df['pop_mulher'] = df['pop_mulher'].astype('uint64')
df['pop_pad_mulher'] = df['pop_pad_mulher'].astype('uint64')
df['mes'] = df['mes'].astype('uint8')
df['notificacoes_total'] = df['notificacoes_total'].astype('uint32')
df['notificacoes_homem'] = df['notificacoes_homem'].astype('uint32')
df['notificacoes_mulher'] = df['notificacoes_mulher'].astype('uint32')
df['grupo'] = df['grupo'].astype('uint8')

#df.to_hdf("pre_taxa_sih.h5", "df")
df.info()
```

```python
df
```

```python
df_sf = df.groupby(['municipio', 'id_municipio', 'ano', 'mes', 'agravo', 'grupo'], as_index=False).sum()
```

```python
df_sf['taxa_notific_total'] = (df_sf['notificacoes_total'] / df_sf['pop_total']) * 100000
df_sf['taxa_notific_homem'] = (df_sf['notificacoes_homem'] / df_sf['pop_homem']) * 100000
df_sf['taxa_notific_mulher'] = (df_sf['notificacoes_mulher'] / df_sf['pop_mulher']) * 100000
```

```python
df_sf.to_csv('taxa_sinan.csv', index=False)
```
