```python
import pandas as pd
import numpy as np

df = pd.read_csv("pre_taxa_sim.csv")
df
```

```python
df['id_municipio'] = df['id_municipio'].astype('uint32')
df['ano'] = df['ano'].astype('uint16')
df['pop_total'] = df['pop_total'].astype('uint64')
df['pop_pad_total'] = df['pop_pad_total'].astype('uint64')
df['pop_homem'] = df['pop_homem'].astype('uint64')
df['pop_pad_homem'] = df['pop_pad_homem'].astype('uint64')
df['pop_mulher'] = df['pop_mulher'].astype('uint64')
df['pop_pad_mulher'] = df['pop_pad_mulher'].astype('uint64')
df['mes'] = df['mes'].astype('uint8')
df['obitos_total'] = df['obitos_total'].astype('uint32')
df['obitos_homem'] = df['obitos_homem'].astype('uint32')
df['obitos_mulher'] = df['obitos_mulher'].astype('uint32')
df['taxa_obitos_total'] = df['taxa_obitos_total'].astype('float32')
df['taxa_obitos_homem'] = df['taxa_obitos_homem'].astype('float32')
df['taxa_obitos_mulher'] = df['taxa_obitos_mulher'].astype('float32')
df['obitos_esp_total'] = df['obitos_esp_total'].astype('float32')
df['obitos_esp_homem'] = df['obitos_esp_homem'].astype('float32')
df['obitos_esp_mulher'] = df['obitos_esp_mulher'].astype('float32')
df['grupo'] = df['grupo'].astype('uint8')
df['taxa_obitos_esp_total'] = df['taxa_obitos_esp_total'].astype('float32')
df['taxa_obitos_esp_homem'] = df['taxa_obitos_esp_homem'].astype('float32')
df['taxa_obitos_esp_mulher'] = df['taxa_obitos_esp_mulher'].astype('float32')

df.to_hdf("pre_taxa_sim.h5", "df")
df.info()
```

```python
df_sf = df.groupby(['municipio', 'id_municipio', 'ano', 'mes', 'agravo', 'grupo'], as_index=False).sum()
```

```python
df_sf['taxa_obitos_total'] = (df_sf['obitos_total'] / df_sf['pop_total']) * 100000
df_sf['taxa_obitos_homem'] = (df_sf['obitos_homem'] / df_sf['pop_homem']) * 100000
df_sf['taxa_obitos_mulher'] = (df_sf['obitos_mulher'] / df_sf['pop_mulher']) * 100000

df_sf['taxa_obitos_esp_total'] = (df_sf['obitos_esp_total'] / df_sf['pop_pad_total']) * 100000
df_sf['taxa_obitos_esp_homem'] = (df_sf['obitos_esp_homem'] / df_sf['pop_pad_homem']) * 100000
df_sf['taxa_obitos_esp_mulher'] = (df_sf['obitos_esp_mulher'] / df_sf['pop_pad_mulher']) * 100000
```
