```python
import pandas as pd
import numpy as np
import datetime

sim = pd.read_csv("sim.csv", encoding='latin-1')
sim = sim.drop(labels=['Unnamed: 0', 'X'], axis=1)
sim['subgrupo'] = np.nan

sim['semana'] = np.nan
sim['dataobito'] = sim['dataobito'].apply(lambda x : datetime.datetime(int(x[0:4]), int(x[5:7]), int(x[8:10])))


df = pd.read_csv("Planilha para o R - Planilha.csv")
df = df.drop(labels='Unnamed: 0', axis=1)
df = df.fillna(0)
df.columns = ['cidsize', 'cid', 'agravo', 'subgrupo']
sim
```

```python
df
```

```python
np.sort(df['agravo'].unique())
```

```python
#itera por cada linha da planilha
for row in df.itertuples():
    #se o código cid tem 3 dígitos verificamos na coluna causa3
    
    if row[1] == 3:
        sim.loc[(sim['causa3'] == row[2]), 'agravo'] = row[3]
        
        #verifica se a doença pertence a algum subgrupo
        if row[4] != 0:
            sim.loc[(sim['causa3'] == row[2]), 'subgrupo'] = row[4]
    
    #se o código cid tem 4 dígitos verificamos na coluna CAUSABAS
    elif row[1] == 4:
        sim.loc[(sim['CAUSABAS'] == row[2]), 'agravo'] = row[3]
        
        #verifica se a doença pertence a algum subgrupo
        if row[4] != 0:
            sim.loc[(sim['CAUSABAS'] == row[2]), 'subgrupo'] = row[4]
```

```python
for row in df.itertuples():
    sim.loc[(sim['agravo'] == row[1]), 'capitulo'] = row[2]
    sim.loc[(sim['subgrupo'] == row[1]), 'capitulo'] = row[2]
```

```python
df = pd.read_csv("semanas.csv")
df
```

```python
def fix_date(data):
    if len(data) < 8:
        data = "0" + data
    
    return data

df['comeco'] = df['comeco'].apply(str)
df['comeco'] = df['comeco'].apply(fix_date)
df['comeco'] = df['comeco'].apply(lambda x : datetime.datetime(int(x[4:8]), int(x[2:4]), int(x[0:2])))

df['fim'] = df['fim'].apply(str)
df['fim'] = df['fim'].apply(fix_date)
df['fim'] = df['fim'].apply(lambda x : datetime.datetime(int(x[4:8]), int(x[2:4]), int(x[0:2])))

df
```

```python
for row in df.iterrows():
    sim.loc[( (sim['dataobito'] >= row[1]['comeco']) 
             & (sim['dataobito'] <= row[1]['fim']) 
             & (sim['ano'] == row[1]['ano']) ), 'semana'] = row[1]['semana']
```
