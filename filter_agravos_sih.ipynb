```python
sih.loc[:,'subgrupo'] = np.nan
sih.loc[:,'semana'] = np.nan
sih
```

```python
df = pd.read_csv("Planilha para o R - SIH - Planilha.csv")
df = df.drop(labels='Unnamed: 0', axis=1)
df = df.fillna(0)
df.columns = ['cidsize', 'cid', 'agravo', 'subgrupo']
df
```

```python
#itera por cada linha da planilha
for row in df.itertuples():
    #se o código cid tem 3 dígitos verificamos na coluna causa3
    if row[1] == 3:
        sih.loc[(sih['causa3'] == row[2]), 'agravo'] = row[3]
        
        #verifica se a doença pertence a algum subgrupo
        if row[4] != 0:
            sih.loc[(sih['causa3'] == row[2]), 'subgrupo'] = row[4]
    
    #se o código cid tem 4 dígitos verificamos na coluna DIAG_PRINC
    elif row[1] == 4:
        sih.loc[(sih['DIAG_PRINC'] == row[2]), 'agravo'] = row[3]
        
        #verifica se a doença pertence a algum subgrupo
        if row[4] != 0:
            sih.loc[(sih['DIAG_PRINC'] == row[2]), 'subgrupo'] = row[4]
```

```python
df = pd.read_csv("planilha - capitulos_sih.csv")
df
```

```python
for row in df.itertuples():
    sih.loc[(sih['agravo'] == row[1]), 'capitulo'] = row[2]
    sih.loc[(sih['subgrupo'] == row[1]), 'capitulo'] = row[2]
```

```python
sih.to_csv("sih_agravos.csv", index=False)
```
