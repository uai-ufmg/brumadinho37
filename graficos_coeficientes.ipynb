```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt

coefs = pd.read_csv("coeficientes.csv")
coefs
```

```python
df_r = coefs[coefs['ano'] < 2019]
coefs_r = df_r.groupby(['grupo'], as_index=False).sum()

coefs_r[['grupo', 'obitos_<1ano', 'obitos_0a27dias', 'obitos_<=6dias', 'obitos_7a27dias', 'obitos_28a364dias', 'obitos_fetais', 'obitos_maternos']]
```

```python
df_19 = coefs[coefs['ano'] == 2019]
coefs_19 = df_19.groupby(['grupo'], as_index=False).sum()
coefs_19

coefs_19[['grupo', 'obitos_<1ano', 'obitos_0a27dias', 'obitos_<=6dias', 'obitos_7a27dias', 'obitos_28a364dias', 'obitos_fetais', 'obitos_maternos']]
```

```python
def calcula_taxas(coefs):
    coefs['tx_mort_infantil'] = ( coefs['obitos_<1ano'] / coefs['nascimentos'] ) * 1000

    coefs['tx_mort_neonatal'] = ( coefs['obitos_0a27dias'] / coefs['nascimentos'] ) * 1000

    coefs['tx_mort_neonatalpre'] = ( coefs['obitos_<=6dias'] / coefs['nascimentos'] ) * 1000

    coefs['tx_mort_neonataltardia'] = ( coefs['obitos_7a27dias'] / coefs['nascimentos'] ) * 1000

    coefs['tx_mort_neonatalpos'] = ( coefs['obitos_28a364dias'] / coefs['nascimentos'] ) * 1000

    coefs['tx_mort_perinatal'] = (( coefs['obitos_<=6dias']  + coefs['obitos_fetais'] ) / coefs['nascimentos'] ) * 1000

    coefs['razao_mort_materna'] = (coefs['obitos_maternos'] / coefs['nascimentos']) * 100000

    capitulos = [ 'I', 'IV', 'V', 'VII', 'IX', 'X', 'XII', 'XIV', 'XVI', 'XX', 'Outros' ]
    for cap in capitulos:
        coef_name = "mort_prop_cap" + cap
        obitos_name = "obitos_cap" + cap
    
        coefs[coef_name] = (coefs[obitos_name] / coefs['obitos']) * 100

    return coefs

coefs_r = calcula_taxas(coefs_r)
coefs_19 = calcula_taxas(coefs_19)
```

```python
coefs_r.columns
```

```python
aux = pd.melt(coefs_r, id_vars=['grupo'], value_vars=['tx_mort_infantil',
       'tx_mort_neonatal', 'tx_mort_neonatalpre', 'tx_mort_neonataltardia',
       'tx_mort_neonatalpos', 'tx_mort_perinatal', 'razao_mort_materna'], var_name='nome_taxa')

aux2 = pd.melt(coefs_19, id_vars=['grupo'], value_vars=['tx_mort_infantil',
       'tx_mort_neonatal', 'tx_mort_neonatalpre', 'tx_mort_neonataltardia',
       'tx_mort_neonatalpos', 'tx_mort_perinatal', 'razao_mort_materna'], var_name='nome_taxa')

aux2['value'] = aux2['value'] - aux['value']
```

```python
c = alt.Chart(aux2, title='Diferença entre as taxas de mortalidade pós e pré rompimento').mark_bar().encode(
    alt.Color('grupo:N', title='Grupo'),
    alt.Y('nome_taxa:N', title='Taxa', sort=None),
    alt.X('value:Q', title='diferença da taxa'),
    alt.Order('grupo:N', sort='ascending'),
    alt.Tooltip('value:Q')
).properties(
    width=640,
).configure_range(
    category={'scheme': 'set1'}
)

c
```
