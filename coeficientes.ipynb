```python
import pandas as pd

sim = pd.read_csv("sim_agravos.csv")
sim
```

```python
sinasc = pd.read_csv("sinasc.csv")
sinasc = sinasc.drop(labels=['Unnamed: 0'], axis=1)
sinasc
```

```python
fetais = pd.read_csv("fetais.csv")
fetais = fetais.drop(labels=['Unnamed: 0', 'X'], axis=1)

fetais
```

```python
aux = sinasc[['CODMUNRES', 'ano', 'IDADEMAE']].groupby(['CODMUNRES', 'ano'], as_index=False).count()
aux.columns = ['CODMUNRES', 'ano', 'nascimentos']
aux
```

```python
obitos = sim[['CODMUNRES', 'ano', 'IDADE']].groupby(['CODMUNRES', 'ano'], as_index=False).count()
obitos.columns = ['CODMUNRES', 'ano', 'obitos']
obitos
```

```python
def get_obitos_dias(sim, min_dias, max_dias, nome_col):
    result = sim[ (sim['dias'] >= min_dias) & (sim['dias'] <= max_dias) ]
    result = result[['CODMUNRES', 'ano', 'IDADE']].groupby(['CODMUNRES', 'ano'], as_index=False).count()
    result.columns = ['CODMUNRES', 'ano', nome_col]
    
    return result

def get_obitos_cap(sim, cap, nome_col):
    result = sim[sim['capitulo'] == cap]
    result = result[['CODMUNRES', 'ano', 'IDADE']].groupby(['CODMUNRES', 'ano'], as_index=False).count()
    result.columns = ['CODMUNRES', 'ano', nome_col]
    
    return result
```

```python
#usado para o coeficiente de mortalidade infantil
obitos_1ano = get_obitos_dias(sim, 0, 364, 'obitos_<1ano')

#usado para o coeficiente de mortalidade neonatal precoce
obitos_6dias = get_obitos_dias(sim, 0, 6, 'obitos_<=6dias')

#usado para o coeficiente de mortalidade neonatal tardia
obitos_7a27dias = get_obitos_dias(sim, 7, 27, 'obitos_7a27dias')

#usado para o coeficiente de Mortalidade Infantil Precoce (ou Neonatal) - CMIP
obitos_0a27dias = get_obitos_dias(sim, 0, 27, 'obitos_0a27dias')

#usado para o coeficiente de mortalidade pós-neonatal
obitos_28a364dias = get_obitos_dias(sim, 28, 364, 'obitos_28a364dias')

#usado para o coeficiente de mortalidade perinatal
obitos_fetais = fetais[['CODMUNRES', 'ano', 'codmunres']].groupby(['CODMUNRES', 'ano'], as_index=False).count()
obitos_fetais.columns = ['CODMUNRES', 'ano', 'obitos_fetais']

#usados para a proporção de mortalidade por causas (capitulo)

obitos_capI = get_obitos_cap(sim, 'I', 'obitos_capI')

obitos_capIV = get_obitos_cap(sim, 'IV', 'obitos_capIV')

obitos_capV = get_obitos_cap(sim, 'V', 'obitos_capV')

obitos_capVII = get_obitos_cap(sim, 'VII', 'obitos_capVII')

obitos_capIX = get_obitos_cap(sim, 'IX', 'obitos_capIX')

obitos_capX = get_obitos_cap(sim, 'X', 'obitos_capX')

obitos_capXII = get_obitos_cap(sim, 'XII', 'obitos_capXII')

obitos_capXIV = get_obitos_cap(sim, 'XIV', 'obitos_capXIV')

obitos_capXVI = get_obitos_cap(sim, 'XVI', 'obitos_capXVI')

obitos_capXX = get_obitos_cap(sim, 'XX', 'obitos_capXX')

obitos_capOutros = get_obitos_cap(sim, 'Outros', 'obitos_capOutros')
```

```python
#usado para razão de mortalidade materna

causas_maternas = ["O0" + str(i) for i in range(1, 10)]
causas_maternas = causas_maternas + ["O" + str(i) for i in range(10, 100)]

causas_maternas.remove("O96")
causas_maternas.remove("O97")

query = "   (causa3 in @causas_maternas)" + \
        "or (causa3 in ['A34', 'F53', 'M83'] and OBITOPUERP == '1')" + \
        "or (causa3 in ['B20', 'B21', 'B22', 'B23', 'B24', 'D39', 'E23'] and (OBITOPUERP == '1' or OBITOGRAV == '1')) "

obitos_maternos = sim.query(query)
obitos_maternos
```

```python
obitos_maternos = obitos_maternos[['CODMUNRES', 'ano', 'IDADE']].groupby(['CODMUNRES', 'ano'], as_index=False).count()
obitos_maternos.columns = ['CODMUNRES', 'ano', 'obitos_maternos']
obitos_maternos
```

```python
coefs = ( aux.merge(obitos_1ano, how='left')
         .merge(obitos_6dias, how='left')
         .merge(obitos_7a27dias, how='left')
         .merge(obitos_0a27dias, how='left')
         .merge(obitos_28a364dias, how='left')
         .merge(obitos_fetais, how='left')
         .merge(obitos_capI, how='left')
         .merge(obitos_capIV, how='left')
         .merge(obitos_capV, how='left')
         .merge(obitos_capVII, how='left')
         .merge(obitos_capIX, how='left')
         .merge(obitos_capX, how='left')
         .merge(obitos_capXII, how='left')
         .merge(obitos_capXIV, how='left')
         .merge(obitos_capXVI, how='left')
         .merge(obitos_capXX, how='left')
         .merge(obitos_capOutros, how='left')
         .merge(obitos_maternos, how='left')
         .merge(obitos, how='left'))

coefs.fillna(0, inplace=True)
coefs
```

```python
coefs['tx_mort_infantil'] = ( coefs['obitos_<1ano'] / coefs['nascimentos'] ) * 1000

coefs['tx_mort_neonatal'] = ( coefs['obitos_0a27dias'] / coefs['nascimentos'] ) * 1000

coefs['tx_mort_neonatalpre'] = ( coefs['obitos_<=6dias'] / coefs['nascimentos'] ) * 1000

coefs['tx_mort_neonataltardia'] = ( coefs['obitos_7a27dias'] / coefs['nascimentos'] ) * 1000

coefs['tx_mort_neonatalpos'] = ( coefs['obitos_28a364dias'] / coefs['nascimentos'] ) * 1000

coefs['tx_mort_perinatal'] = (( coefs['obitos_<=6dias']  + coefs['obitos_fetais'] ) / coefs['nascimentos'] ) * 1000

coefs['razao_mort_materna'] = (coefs['obitos_maternos'] / coefs['nascimentos']) * 100000

capitulos = [ 'I', 'IV', 'V', 'VII', 'IX', 'X', 'XII', 'XIV', 'XVI', 'XX', 'Outros' ]
for cap in capitulos:
    coef_name = "mort_prop_cap" + cap
    obitos_name = "obitos_cap" + cap
    
    coefs[coef_name] = (coefs[obitos_name] / coefs['obitos']) * 100

coefs
```
