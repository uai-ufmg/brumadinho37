```python
def ler_base(nome, encoding=True, sep=";"):
    if encoding:
        df = pd.read_csv(f"bases_socio/{nome}.csv", sep=sep, encoding='latin')
    else:
        df = pd.read_csv(f"bases_socio/{nome}.csv", sep=sep)
        
    return df

def tratar_base(nome, encoding=True, sep=";"):
    df = ler_base(nome, encoding = encoding, sep = sep)
    
    novas_colunas = []
    for i in df.columns:
        if i.isnumeric():
            df[i] = df[i].str.replace(",", ".").astype('float32')
            novas_colunas.append(f"{nome} - {i}")
        else:
            novas_colunas.append(i)
    
    df.columns = novas_colunas
    
    return df

def tratar_base_jp(nome, encoding=True, sep=";"):
    df = ler_base(nome, encoding = encoding, sep = sep)
    
    novas_colunas = []
    for i in df.columns:
        if i.isnumeric():
            df[i] = df[i].str.replace(",", ".").astype('float32')
            novas_colunas.append(f"{nome} - {i}")
        else:
            novas_colunas.append(i)
    
    df.columns = novas_colunas
    
    df['id_municipio'] = df['Município'].str[:6].astype('int64') #coluna do codigo do municipio
    df['Município'] = df['Município'].str[7:] #coluna do nome do municipio
    
    #df.drop('Município', axis=1, inplace=True)
    
    #df = df[df['municipio'].isin(todas)]
    
    return df



df1 = tratar_base(nome1)
df2 = tratar_base(nome2)
df3 = tratar_base(nome3)
df4 = tratar_base(nome4)
df5 = tratar_base(nome5)
df6 = tratar_base(nome6)
df7 = tratar_base(nome7)
df8 = tratar_base(nome8, False)
df8[" Salário médio mensal dos trabalhadores formais"] = df8[" Salário médio mensal dos trabalhadores formais"].str.replace(",", ".").astype('float32')
df8.columns = ["Município", "Salário médio mensal dos trabalhadores formais"]
df9 = tratar_base(nome9)

df10 = tratar_base_jp(nome10)
df11 = tratar_base_jp(nome11)
df12 = tratar_base_jp(nome12)
df13 = tratar_base(nome13, sep=",", encoding = False)
df13.columns = ['Município', 'Índice de Gini 2010', 'IDHM 2010']
df14 = tratar_base_jp(nome14)
df15 = tratar_base_jp(nome15)
df16 = tratar_base_jp(nome16)
df17 = tratar_base_jp(nome17)
df18 = tratar_base_jp(nome18)
df19 = tratar_base_jp(nome19)
```
